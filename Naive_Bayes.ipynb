```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
spam_data=pd.read_csv('spam_ham.csv',encoding="ISO-8859-1")
spam_data
```

```python
spam_data.dropna(axis=1,inplace=True)
spam_data.dropna(axis=0,inplace=True)
```

```python
spam_data
```

```python
spam_data['v1']=spam_data.v1.replace({'spam':1,'ham':0})
spam_data['v2']=spam_data['v2'].str.replace('\W',' ')
spam_data['v2']=spam_data['v2'].str.lower()
spam_data
```

```python
spam_data['v2']=spam_data['v2'].str.replace('.',' ')
spam_data['v2']=spam_data['v2'].str.replace('_',' ')
spam_data
```

```python
spam_count=spam_data.groupby('v1')['v1'].count().reset_index(name='Counts')
spam_count.index=['ham','spam']
spam_count
```

```python
np.random.seed(0)
df_train,df_test=train_test_split(spam_data,train_size=0.7,test_size=0.3,random_state=100)
y_train=df_train.pop('v1')
x_train=df_train
y_test=df_test.pop('v1')
x_test=df_test
print(x_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer(stop_words='english',analyzer='word')
x_train_list=x_train['v2'].tolist()
print(x_train_list)
vector.fit(x_train_list)

```

```python
vector.vocabulary_
```

```python
x_train_transformed=vector.transform(x_train['v2'])
x_test_transformed=vector.transform(x_test['v2'])
```

```python
mnb=MultinomialNB()
mnb.fit(x_train_transformed,y_train)
y_pred=mnb.predict(x_train_transformed)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,y_pred)
```

```python
y_test_pred=mnb.predict(x_test_transformed)
accuracy_score(y_test,y_test_pred)
```
